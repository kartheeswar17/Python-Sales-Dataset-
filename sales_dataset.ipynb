```python
import numpy as np
n = int(input("Number of products: "))
qty = [float(input(f"Qty {i+1}: ")) for i in range(n)]
prc = [float(input(f"Price {i+1}: ")) for i in range(n)]
q_arr, p_arr = np.array(qty), np.array(prc)
sales = q_arr * p_arr
print(f"\nSales per product: {sales}")
print(f"Top 3: {sales[:3]}")
print("\n--- Stats ---")
print(f"Total:   {sales.sum():.2f}")
print(f"Average: {sales.mean():.2f}")
print(f"Median:  {np.median(sales):.2f}")
print(f"Std Dev: {sales.std():.2f}")
```
